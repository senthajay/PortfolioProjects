```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\DataAnalyst_Projects\Pandas_projects_Datasets\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
```

```python
df
```

```python
df[["Street Address", "State","Postal Code"]]=df["Address"].str.split(',', n=2, expand = True)
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].astype(str).str.replace(r'\D', '', regex=True)
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: '-'.join([x[:3],x[3:6],x[6:]]))
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].astype(str).str.replace('--', '', regex=True)
```

```python
df
```

```python
# remove the not nescessray column 

df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes","Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No","N")
```

```python
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna("") # to remove the null or missing values 
```

```python
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes","Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No","N")
df
```

```python
df
```

```python
df = df.replace("N/a", "")
```

```python
df = df.fillna(" ")
```

```python
df
```

```python
# to give only the list of contacts that are allowed to be contacted 

for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == "Y":
        df.drop(x, inplace = True)
        
df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == "":
        df.drop(x, inplace = True)
        
df
```

```python
df["Paying Customer"]=df["Paying Customer"].replace("Yes","Y")
```
