```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\DataAnalyst_Projects\Pandas_projects_Datasets\movies.csv")
df
```

```python
# to find the missing values in the data set 
for col in df.columns:
    missing_data = np.sum(df[col].isnull())
    print('{} - {}'.format(col,round(missing_data)))
```

```python
# now that we found there are some data missing, we need to clean the missing data 
df = df.dropna()
```

```python
df.head()
```

```python
# re-check missing data 
for col in df.columns:
    missing_data = np.sum(df[col].isnull())
    print('{} - {}'.format(col,round(missing_data)))
```

```python
df.head()
```

```python
# now we check the year and released year to be matched, if not matched update the year to the released year.
#check the data type
df.dtypes
```

```python
# convert the released col to str and extract the year to creat a new col
df['year_correct'] = df['released'].str.extract(pat = '([0-9]{4})').astype('int64')
```

```python
df.head()
```

```python
corrected_y = df['year'] == df['year_correct']
sum(corrected_y)
```

```python
mismatched_y = ~corrected_y
sum(mismatched_y)
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
df_numerized = df.copy()


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
# using sns heatmap analysis to find the corr

correlation_matrix = df_numerized.corr(numeric_only = True)
sns.heatmap(correlation_matrix, annot = True)


plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show
```

```python
# to view correlation between pairs
corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack()

corr_pairs
```

```python
#sorting correlated pairs
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#highly correlated pairs
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
